```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

FILE_NAME = "parameter_sweep_results.csv"
SAVE_DIR = '/Users/Chiara/Documents/Documents_Chiara_MacBook_Air/Education : Personal Projects : Studies/Data Driven Modeling/Newtonian-SINDy/GS/GS\W\SINDy'

file_path = os.path.join(SAVE_DIR, FILE_NAME)

headers = ["years",
           "inv_points_per_year",
           "time_taken_s",
           "Sun_solver_jupiter_mass",
           "Mercury_solver_sun_mass",
           "Mercury_solver_jupiter_mass",
           "Venus_solver_sun_mass",
           "Venus_solver_jupiter_mass",
           "Earth_solver_sun_mass",
           "Earth_solver_jupiter_mass",
           "Mars_solver_sun_mass",
           "Mars_solver_jupiter_mass",
           "Jupiter_solver_sun_mass"]

peek = pd.read_csv(file_path, nrows=1, header=None)
has_header = peek.iloc[0, 0] == headers[0]
df = pd.read_csv(file_path, header=0 if has_header else None, names=headers)
df = df.apply(pd.to_numeric, errors='coerce')

# print(df.dtypes)
# print(df.head())

for h in headers:
    if h in ("years", "inv_points_per_year", "time_taken_s"):
        continue
    df[h] = (df[h] / (df["years"]**2)) * (1 / 3.154e7)**2 * 1.496e11**3

output_path = os.path.join(SAVE_DIR, "updated_" + FILE_NAME)
df.to_csv(output_path, index=False)

print(f"Saved to {output_path}")
```
